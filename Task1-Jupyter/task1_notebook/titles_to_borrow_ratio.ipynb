```python
import pandas as pd

df = pd.read_csv('PDV5_01-Task_1-books.csv')

df.head()
```

Am încărcat datele și am verificat primele rânduri pentru a ne asigura că fișierul a fost citit corect. Ne interesează în special coloanele `section`, `title` și `times_borrowed`.


## 🧮 Calculul raportului de dezechilibru

Vrem să vedem:
- câte titluri unice există pe fiecare secțiune
- de câte ori au fost împrumutate cărțile din fiecare secțiune

Astfel putem calcula un **raport cerere / ofertă**, unde:


```python
# Grupăm datele pe secțiune:
# - numărăm câte titluri există
# - totalizăm câte împrumuturi au fost
inventory_gap = df.groupby('section').agg({
    'title': 'count',
    'times_borrowed': 'sum'
})

# Calculăm raportul dintre împrumuturi și titluri
inventory_gap['titles_to_borrow_ratio'] = (
    inventory_gap['times_borrowed'] / inventory_gap['title']
)

# Sortăm descrescător după acest raport
inventory_gap_sorted = inventory_gap.sort_values(
    by='titles_to_borrow_ratio',
    ascending=False
)

# Afișăm rezultatele
inventory_gap_sorted
```

## ✅ Rezumatul rezultatului

Secțiunea cu cel mai mare **raport cerere / ofertă** este listată prima în tabelul de mai sus.  
Aceasta înseamnă că în acea secțiune:
- sunt relativ puține titluri
- dar sunt împrumutate de foarte multe ori

Acest lucru ne indică **un dezechilibru de inventar**: publicul cere mai multe titluri decât avem în ofertă pe acea tematică.

Rezultatul ne sugerează că secțiunea cu cel mai mare dezechilibru ar trebui prioritar extinsă prin achiziții suplimentare.
